```python
import warnings
#importing libraries for tabular data and plotting
import hvplot.pandas
import pandas as pd

warnings.filterwarnings('ignore')
```

```python
# Define the URL for LA mean temp data download
la_temp_url = (
    'https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/'
    'city/time-series/USW00026451/tmax/1/8/1895-2023.csv?base_prd=true'
    '&begbaseyear=1926&endbaseyear=2022')
```

```python
# Turning data URL into a workable dataframe
LA_temp_DF = pd.read_csv(la_temp_url, header=4, 
                         names=['year', 'temp_f'],
                         usecols=[i for i in range(2)],)
```

```python
#extracting year
LA_temp_DF.year = pd.to_datetime(LA_temp_DF.year, format='%Y%m').dt.year
```

```python
LA_temp_DF.plot(
    x='year', y='temp_f', 
    title = 'Max Temperature in Achorage, AK, USA from 1952-2023', 
    color = "red",
    xlabel = 'Year', ylabel = 'Temperature (*F)')
```
